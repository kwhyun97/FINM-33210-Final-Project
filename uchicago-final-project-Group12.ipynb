## Data Cleaning and Winsorization

The distribution of many statistics can be heavily influenced by outliers. A simple approach to robustifying parameter estimation procedures is to set all outliers to a specified percentile of the data; for example, a 90% winsorization would see all data below the 5th percentile set to the 5th percentile, and data above the 95th percentile set to the 95th percentile. Winsorized estimators are usually more robust to outliers than their more standard forms.

```python
def wins(x,a,b):
    return(np.where(x <= a,a, np.where(x >= b, b, x)))

def clean_nas(df): 
    numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
    
    for numeric_column in numeric_columns: 
        df[numeric_column] = np.nan_to_num(df[numeric_column])
    
    return df
```

We can check the distribution of returns with a density plot, both before and after winsorization to observe the effect of trimming the tails.

```python
def density_plot(data, title): 
    density = gaussian_kde(data)
    xs = np.linspace(np.min(data),np.max(data),200)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs))
    plt.title(title)
    plt.show()
    
test = frames['20040102']
density_plot(test['Ret'], 'Daily return pre-winsorization')
density_plot(wins(test['Ret'],-0.2,0.2), 'Daily return winsorized')

D = (test['SpecRisk'] / (100 * math.sqrt(252))) ** 2
density_plot(np.sqrt(D), 'SpecRisk')
```

## Factor Exposures and Factor Returns


Arbitrage pricing theory relaxes several of the assumptions made in the course of deriving 
the CAPM. In particular, we relax the assumption that all investors do the same optimization 
and hence that there is a single efficient fund. This allows the possibility that a CAPM-like 
relation may hold, but with multiple underlying sources of risk. 

Specifically, let $r_i, i = 1, \ldots, n$ denote the cross-section of asset returns over a given 
time period $[t, t+1]$. In a fully-general model, the multivariate distribution $p(\mathbf{r})$ could have 
arbitrary covariance and higher-moment structures, but remember that for $n$ large there is 
typically never enough data to estimate such over-parameterized models. 

Instead, we assume a
structural model which is the most direct generalization of the CAPM: 
$$
	r_i = \beta_{i, 1} f_1 + \beta_{i,2} f_2 + \dots + \beta_{i, p} f_p + \epsilon_i, \quad 
	\epsilon_i \sim N(0, \sigma_i^2)  
$$
If $p = 1$, this reduces to the Capital Asset Pricing Model (CAPM) in a rather direct way. 

With $p > 1$, the model starts to differ from the CAPM 
in several very important aspects. In the CAPM, we were 
able to identify the single efficient fund by arguing that its weights must equal the 
market-capitalization weights. Hence we were given for free a very nice proxy for the 
single efficient fund: a capitalization-weighted basket such as the Russell 3000. 
Hence in the $p=1$ case we had a convenient proxy which could be used to impute 
the return $f_1$, which we called $r_M$. Also $\beta_{i, 1}$ could be estimated, with 
no more than the usual statistical estimation error, by time-series regression. 

If $p > 1$ then the underlying assumptions of that argument break down: there is no
longer any simple way to identify $f_{j}$ nor $\beta_{i, j}$ ($j = 1, \ldots, p$). 
We shall return to the estimation problem in due course. 


To avoid confusion with the CAPM, and its simplistic $\beta$ coefficient (which is still 
sometimes used in larger multi-factor models), it is conventional to make the following 
notation change: $\beta_{i,j}$ becomes $X_{i,j}$ and so the model equation becomes 
$$
	 	r_i = X_{i, 1} f_1 + X_{i,2} f_2 + \dots + X_{i, p} f_p + \epsilon_i, \quad 
	\epsilon_i \sim N(0, \sigma_i^2)  
$$
It's difficult to simultaneously estimate both all components $X_{i,j}$ and all risk-source 
returns $f_j$, so one usually assumes one is known and calculates the other via regression. 
In what follows, we focus on the approach where $X$ is known, and the $f_j$ are assumed
to be hidden (aka latent) variables.

The structural equation is more conveniently expressed in matrix form: 
$$
	R_{t+1} = X_t f_{t+1} + \epsilon_{t+1}, \quad E[\epsilon] = 0, \ \ V[\epsilon] = D 
$$
where $R_{t+1}$ is an $n$-dimensional random vector containing the 
cross-section of returns in excess of the risk-free rate over some 
time interval $[t, t+1]$, and $X_t$ is a (non-random) $n \times p$ matrix 
that can be calculated entirely from data known before time $t$.
The variable $f$ in denotes a $p$-dimensional random vector
process which cannot be observed directly. 

Since the variable $f$ denotes a $p$-dimensional random vector process which cannot 
be observed directly, information about the $f$-process must be obtained via statistical inference. 
We assume that the $f$-process has finite first and second moments given by 
$$
	E[f] = \mu_f, \ \text{ and } \  V[f] = F . 
$$ 
The primary outputs of a statistical inference process are the parameters 
$\mu_f$ and $F$, and other outputs one might be interested in
include estimates of the daily realizations $\hat{f}_{t+1}$. 

The simplest way of estimating historical daily realizations of $\hat{f}_{t+1}$ is by least-squares (ordinary or weighted, as appropriate), viewing the defining 
model equation as a regression problem. 

```python
industry_factors = ['AERODEF', 'AIRLINES', 'ALUMSTEL', 'APPAREL', 'AUTO',
       'BANKS','BEVTOB', 'BIOLIFE', 'BLDGPROD','CHEM', 'CNSTENG', 'CNSTMACH', 'CNSTMATL', 'COMMEQP', 'COMPELEC',
       'COMSVCS', 'CONGLOM', 'CONTAINR', 'DISTRIB',
       'DIVFIN', 'ELECEQP', 'ELECUTIL', 'FOODPROD', 'FOODRET', 'GASUTIL',
       'HLTHEQP', 'HLTHSVCS', 'HOMEBLDG', 'HOUSEDUR','INDMACH', 'INSURNCE', 'INTERNET', 
        'LEISPROD', 'LEISSVCS', 'LIFEINS', 'MEDIA', 'MGDHLTH','MULTUTIL',
       'OILGSCON', 'OILGSDRL', 'OILGSEQP', 'OILGSEXP', 'PAPER', 'PHARMA',
       'PRECMTLS','PSNLPROD','REALEST',
       'RESTAUR', 'ROADRAIL','SEMICOND', 'SEMIEQP','SOFTWARE', 'SPLTYRET', 'SPTYCHEM', 'SPTYSTOR',
       'TELECOM', 'TRADECO', 'TRANSPRT', 'WIRELESS']

style_factors = ['BETA', 'SIZE', 'MOMENTUM', 'VALUE', 'GROWTH', 'LEVERAGE', 'LIQUIDTY', 
                 'DIVYILD', 'LTREVRSL', 'EARNQLTY']

## an R-style formula which can be used to construct a cross sectional regression
def get_formula(alphas, Y):
    L = ["0"]
    L.extend(alphas)
    L.extend(style_factors)
    L.extend(industry_factors)
    return Y + " ~ " + " + ".join(L)

## The term 'estu' is short for estimation universe
def get_estu(df): 
    estu = df.loc[df.IssuerMarketCap > 1e9].copy(deep=True)
    return estu

def estimate_factor_returns(df, alphas): 
    ## build universe based on filters 
    estu = get_estu(df)
    
    ## winsorize returns for fitting 
    estu['Ret'] = wins(estu['Ret'], -0.25, 0.25)
    
    model = ols(get_formula(alphas, "Ret"), data=estu)
    return(model.fit())
```

## 1 - 1. Composite Alpha Factor Models

For each individual equity $i$, the current model we have with the 6 alpha factors is:

$$
r_{i, t} = X_{i, 1} f_{1, t}^{\alpha} + X_{i, 2} f_{2, t}^{\alpha} + \dots + X_{i, 6} f_{6, t}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_t +
\sum X_{i, Style} \text{Style Factors}_t +
\epsilon_{i, t}
$$

What we are trying to achieve is a combination of the six alpha factors that describe the variation
in returns for each equity $i$ ($r_{i, t}$).

If we let $g(f_1, f_2, \dots, f_6, t)$ to be that combination of the six factors,
mathematically our goal can be expressed as:

$$
r_{i, t} = \beta_i^{composite} \cdot g(f_1, f_2, \dots, f_6, t) + \epsilon_{i, t}
$$

We have tried multiple ways to attain such $g(f_1, f_2, \dots, f_6, t)$ that predicts the data the best.

### 1 - 1 - 1. Polynomial Model (LASSO/Ridge)

The first method we are going to consider is LASSO $l_1$ optimization on the polynomial model.
This method was suggested under the naive assumption that $g(f_1, f_2, \dots, f_6, t)$ can be
prescribed with linear/non-linear combinations of $f_i$.

For instance, if we let

$$
\begin{aligned}
g_1(f_1, f_2, \dots, f_6, t) &= a_1^{[1]} f_{1, t} + a_2^{[1]} f_{2, t} + \dots a_6^{[1]} f_{6, t} \\
g_2(f_1, f_2, \dots, f_6, t) &= a_1^{[2]} f_{1, t}^2 + a_2^{[2]} f_{2, t}^2 + \dots a_6^{[2]} f_{6, t}^2 +
\sum_{l \neq m} a_{l, m}^{[2]} f_{l, t} f_{m, t} \\
g_3(f_1, f_2, \dots, f_6, t) &= a_1^{[3]} f_{1, t}^3 + a_2^{[3]} f_{2, t}^3 + \dots a_6^{[3]} f_{6, t}^3 +
\sum_{l \neq m} a_{l, m}^{[3]} f_{l, t}^2 f_{m, t} +
\sum_{l \neq m, m \neq n, n \neq l} a_{l, m, n}^{[3]} f_{l, t} f_{m, t} f_{n, t}
\end{aligned}
$$

where $a^{[p]} \in \mathbf{R}$.

With the right $a^{[p]}$ values, the $g(f_1, f_2, \dots, f_6, t)$ could be expressed in the form:

$$
g(f_1, f_2, \dots, f_6, t) \approx g_1(f_1, f_2, \dots, f_6, t) +
g_2(f_1, f_2, \dots, f_6, t) + g_3(f_1, f_2, \dots, f_6, t) + \hat{O}(f_i)^4
$$

Another observation was that from

$$
r_{i, t} = \beta_i^{composite} \cdot g(f_1, f_2, \dots, f_6, t) + \epsilon_{i, t}
$$

We may also deduce

$$
\begin{aligned}
\sum_{i = 1}^N r_{i, t} &= \sum_{i = 1}^N \beta_i^{composite} \cdot g(f_1, f_2, \dots, f_6, t) + \sum_{i = 1}^N
\epsilon_{i, t} \\
&= g(f_1, f_2, \dots, f_6, t) \sum_{i = 1}^N \beta_i^{composite} + \sum_{i = 1}^N \epsilon_{i, t} \\
\Leftrightarrow
\frac{1}{N} \sum_{i = 1}^N r_{i, t} &= g(f_1, f_2, \dots, f_6, t) \frac{1}{N} \sum_{i = 1}^N \beta_i^{composite} +
\frac{1}{N} \sum_{i = 1}^N \epsilon_{i, t} \\
\Leftrightarrow
\mathbf{E}\left[\frac{1}{N} \sum_{i = 1}^N r_{i, t} \right] &=
\left( \frac{1}{N} \sum_{i = 1}^N \beta_i^{composite} \right) \mathbf{E} \left[g(f_1, f_2, \dots, f_6, t) \right] +
\mathbf{E} \left[\frac{1}{N} \sum_{i = 1}^N \epsilon_{i, t}\right] \\
&= \left( \frac{1}{N} \sum_{i = 1}^N \beta_i^{composite} \right) \mathbf{E} \left[g(f_1, f_2, \dots, f_6, t) \right]
\end{aligned}
$$

Given that $\frac{1}{N} \sum_{i = 1}^N \beta_i^{composite}$ is a fixed value across time, if we regress
$f_l$, $f_l^2$, $f_l f_m$, $f_l^3$, $f_l^2 f_m$, $f_l f_m f_n$ against the daily average return
across all available equities in the sample universe,

we would get coefficients that are analogous to
$\left(\frac{1}{N} \sum_{i = 1}^N \beta_i^{composite} \right) a^{[p]} \propto a^{[p]}$,
assuming $\frac{1}{N} \sum_{i = 1}^N \beta_i^{composite} \neq 0$

Therefore, we have performed LASSO regression with $\lambda$ optimized by cross-validation on
$f_l$, $f_l^2$, $f_l f_m$, $f_l^3$, $f_l^2 f_m$, $f_l f_m f_n$ against daily average returns

After doing so, the LASSO coefficients were believed to specify
$g_1(f_1, f_2, \dots, f_6, t)$, $g_2(f_1, f_2, \dots, f_6, t)$, and $g_3(f_1, f_2, \dots, f_6, t)$
which would together form our estimate $\hat{g}(f_1, f_2, \dots, f_6, t)$.

```python
facret_der = facret_df.copy()

# Compute multiplicative factor returns, interaction terms
p = q = range(0, 6)
pq = [(x, y) for x in p for y in q if x <= y]
pq
for i in range(0, len(pq)):
   p = alpha_factors[pq[i][0]]
   q = alpha_factors[pq[i][1]]
   facret_der['{}_{}'.format(p, q)] = facret_der[p] * facret_der[q]

# Compute multiplicative three factor returns
p = q = m = range(0, 6)
pqm = [(x, y, z) for x in p for y in q for z in m if x <= y <= z]
pqm
for i in range(0, len(pq)):
   p = alpha_factors[pqm[i][0]]
   q = alpha_factors[pqm[i][1]]
   m = alpha_factors[pqm[i][2]]
   facret_der['{}_{}_{}'.format(p, q, m)] = facret_der[p] * facret_der[q] * facret_der[m]

facret_der
```

Above shows a new data frame of the alpha factor returns with all possible first-order, second-order, and third-order
combinations (together **48 predictors**).

Since we are dealing with time-series data, we would need to do a rolling window cross validation across the
training set to get the optimal $\lambda$.

```python
lambdas = 10**np.linspace(-3, 10, num=1000)
mse_values = {}
for lam in lambdas:
    mse = []
    for year in [2003, 2004, 2005, 2006]:
        scaler = StandardScaler()
        # dividing train data
        this_year_start = str(year) + "-01-01"
        this_year_end = str(year) + "-12-31"
        X_train = facret_der.loc[this_year_start:this_year_end].values
        X_train_scaled = scaler.fit_transform(X_train)
        y_train = mean_ret_df.loc[this_year_start:this_year_end].values
        y_train_scaled = scaler.fit_transform(y_train)
        # dividing validation data
        next_year_start = str(year+1) + "-01-01"
        next_year_end = str(year+1) + "-12-31"
        X_valid = facret_der.loc[next_year_start:next_year_end].values
        X_valid_scaled = scaler.fit_transform(X_valid)
        y_valid = mean_ret_df.loc[next_year_start:next_year_end].values
        y_valid_scaled = scaler.fit_transform(y_valid)
        # fitting the lasso on train data
        lasso = Lasso(alpha = lam)
        lasso.fit(X_train_scaled, y_train_scaled)
        # validation
        y_pred = lasso.predict(X_valid_scaled)
        mse.append(np.mean((y_pred - y_valid_scaled)**2))
    mse_values[lam] = mse
mse_values = pd.DataFrame(mse_values).transpose()
```

This would mean that for $\lambda$ above a certain value would result in all coefficients forced to $0$ due to the
$l_1$ penalty. In other words, the LASSO model suggests that **polynomial combinations of the six alpha factors
do not explain the variation in mean returns**.

While anticipating low prediction power, we have decided to fit a polynomial model using the minimum $\lambda$
that penalizes all coefficients to $0$ in a Ridge regression. Though the model will not perform well in terms
of predicting tomorrow returns, we believed that having some coefficients would yield to explaining the
variation in the data with polynomial terms.

```python
# getting the smallest hyper-parameter with minimum MSE value
fake_lambda = mse_values.mean(axis=1)[mse_values.mean(axis=1) == mse_values.mean(axis=1).min()].index[0]
ridge = Ridge(alpha = fake_lambda)
ridge.fit(facret_der, mean_ret_df.values)
None
```

From the ridge regression, using the minimum lowest-MSE-yielding $\lambda$ from lasso,
we are able to get our approximation of the "composite alpha factor" as below:

```python
def Poly_Model_composite(six_factors, ridge_coefs):
    facret_der = six_factors.copy()

    # Compute multiplicative factor returns, interaction terms
    p = q = range(0, 6)
    pq = [(x, y) for x in p for y in q if x <= y]
    pq
    for i in range(0, len(pq)):
       p = alpha_factors[pq[i][0]]
       q = alpha_factors[pq[i][1]]
       facret_der['{}_{}'.format(p, q)] = facret_der[p] * facret_der[q]

    # Compute multiplicative three factor returns
    p = q = m = range(0, 6)
    pqm = [(x, y, z) for x in p for y in q for z in m if x <= y <= z]
    pqm
    for i in range(0, len(pq)):
       p = alpha_factors[pqm[i][0]]
       q = alpha_factors[pqm[i][1]]
       m = alpha_factors[pqm[i][2]]
       facret_der['{}_{}_{}'.format(p, q, m)] = facret_der[p] * facret_der[q] * facret_der[m]

    composite_factor = pd.DataFrame(np.dot(facret_der.values, ridge_coefs.T),
                                    index = facret_der.index, columns=["Composite Alpha"])
    return composite_factor
```

### 1 - 1 - 2. PCA Model

Principal Component Analysis (PCA) procedure is to find mutually uncorrelated linear combinations $Y_i$
(principal components) of the original variable $X_j$’s,

$$
Y_i = a'_{i}X = a_{i1}X_1 + · · · + a_{i6}X_6 = \sum_{j = 1}^6  a_{ij}X_j ,  i = 1,... , 6,
$$

such that the variance $var(Y_i), i = 1,... , 6$ is orderly maximized. The hope is that the first principal components
carry most of the variations in the data. Assume that the p-variate vector $X = (X_1, \cdots , X_6) = [X_1 \cdots X_6]'$
has covariance matrix $\sum_x = \sum$ with eigenvalues $\lambda_1 \ge \lambda_2 \ge · · ·\ge \lambda_p \ge 0$. The
method of Lagrange Multipliers is used sequentially to maximize $var(Yi)$ subject to constraints $\sum_{j=1}^n
a^2_{ij} = a'_i a_i = 1$ and $cov(Y_i, Y_k) = 0$ for all $k < i$ (thus for all $k \neq i$).

We choose PCA(n_components = 'mle' , svd_solver='full') (Minka’s MLE is used to guess the dimension) to find the 1
maximal likelihood principal component ($\lambda_1$) that represents the 6 given alpha factors for each day for each
security. Then I applied linear regression between the principal component and ret for that day. The coefficient
represents the composite alpha factor for each day.

$$
r_{i, t} = \beta_{i,t}  \lambda_{i,t} + \epsilon_{i, t}
$$

We tried PCA and TruncatedSVD. Both gave very similar result so I picked to PCA to optimize the composite alpha factor.

```python
def PCA_OLS(frames: pd.DataFrame,
            facret_df: pd.DataFrame,
            alpha_factors: list ):
    PC = []
    # SVD = []
    composite_factors = []
    pred_ret = []

    # for each day
    for d_str, d_td in zip(frames.keys(), list(facret_df.index)):
        estu = get_estu(frames[d_str])
        estu['Ret'] = wins(estu['Ret'], -0.25, 0.25)
        x = (estu.loc[:, alpha_factors] * facret_df.loc[d_td ,]).values
        y = estu.Ret.values
        pca = make_pipeline(StandardScaler(),
                            PCA(n_components = 'mle' , svd_solver='full') )
        # Using Standard Scaler because PCA is highly sensitive to variance in data
        # PCA(n_components = 'mle' , svd_solver='full') use Minka’s MLE to guess the dimension

        principalComponent = pca.fit_transform(x,y)   # composite alpha factor for each security

        # svd = TruncatedSVD(n_components=1, n_iter=10, random_state=0)
        # svd_Component = svd.fit_transform(x,y)

        reg = linear_model.LinearRegression()   # use OLS for simplicity because the data is just 1 independent variable
        composite_factor = reg.fit(principalComponent, y).coef_[0]
        pred = reg.predict(principalComponent)  # 1 composite alpha factor for each day

        PC.append( principalComponent)
        # SVD.append(svd_Component)
        composite_factors.append(composite_factor)
        pred_ret.append(pred.mean())

    return composite_factors, pred_ret
```

### 1 - 1 - 3. Generalized Additive Model (GAM)

The GAM model builds on the idea of regressing functions of $f_1, \dots, f_6$ against the daily average return
of stocks (from here will be noted as $\bar{r}_t$.

What the Generalized Additive Model does is it applies smoothing spline to each of the predictor variables.
In mathematical notations:

$$
\bar{r}_t = \beta_0 + h_1(f_{1, t}) + h_2(f_{2, t}) + \dots + h_6(f_{6, t}) + \epsilon_t
$$

where each smoothing spline $h_i()$ is optimized for each $f_i$ under the constraint

$$
\min_{h_i(x)} \sum_{t = 1}^N(\bar{r}_t - h_i(f_{i, t}))^2 + \lambda \int h_i''(t)^2 dt
$$

and the tuning parameter $\lambda$ is optimized using cross-validation

After we let the GAM algorithm optimize the different smoothing spline, we may make the assumption that
the composite alpha factor we are looking for can be specified by weighting the different smoothing splines as:

$$
\hat{g}(f_1, f_2, \dots, f_6, t) = h_1(f_{1, t}) + h_2(f_{2, t}) + \dots + h_6(f_{i, t})
$$

The downside of GAM is that the optimized smoothing splines $h_i(x)$ are often hard to interpret.
However, since we are exploring various combinations of the alpha factors that best predict the
daily returns, interpretability won't be a hindrance.

After we get our approximated $\hat{g}(f_1, f_2, \dots, f_6, t)$ for each day, we will regress

$$
r_{i, t} = \beta_i^{composite} \cdot \hat{g}(f_1, f_2, \dots, f_6, t) + \epsilon_{i, t}
$$

to get each equity's beta to the composite alpha factor and the error in prediction.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(facret_df.values)
y_scaled = scaler.fit_transform(mean_ret_df.values)
gam = LinearGAM(s(0) + s(1) + s(2) + s(3) + s(4) + s(5), fit_intercept=False)
gam.gridsearch(X_scaled, y_scaled)
None
```

### 1 - 1 - 4. Independent Component Analysis (ICA) - Gradient Boosting Model

The Independent Component Analysis method stems from the fundamental logic behind trying to find the composite
alpha factor. In this given situation, we have one regression model that has been fitted and used for estimation of
the six alpha factors:

$$
r_{i, t} = X_{i, 1} f_{1, t}^{\alpha} + X_{i, 2} f_{2, t}^{\alpha} + \dots + X_{i, 6} f_{6, t}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_t +
\sum X_{i, Style} \text{Style Factors}_t +
\epsilon_{i, t}
$$

We also have a model we hope to attain by combining various alpha factor models:

$$
r_{i, t} = \beta_i^{composite} \cdot g(f_1, f_2, \dots, f_6, t) + \epsilon_{i, t}
$$

Assuming that **both models are true** and that **the error variance in both models are not too big or different**,
we may also be able to state that:

$$
\beta_i^{composite} \cdot g(f_1, f_2, \dots, f_6, t) \sim
X_{i, 1} f_{1, t}^{\alpha} + X_{i, 2} f_{2, t}^{\alpha} + \dots + X_{i, 6} f_{6, t}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_t +
\sum X_{i, Style} \text{Style Factors}_t
$$

This would mean that the values of

$$
X_{i, 1} f_{1, t}^{\alpha} + X_{i, 2} f_{2, t}^{\alpha} + \dots + X_{i, 6} f_{6, t}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_t +
\sum X_{i, Style} \text{Style Factors}_t
$$

can be explained by the product of $\beta_i^{composite}$ that varies across equities but not across time
and $g(f_1, f_2, \dots, f_6, t)$ that varies across time but not across equities.

Therefore, we will conduct ICA to recover two independent signals from the matrix

$$
\begin{bmatrix}
\cdots &
X_{i, 1} f_{1, t_1}^{\alpha} + X_{i, 2} f_{2, t_1}^{\alpha} + \dots + X_{i, 6} f_{6, t_1}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_{t_1} +
\sum X_{i, Style} \text{Style Factors}_{t_1} & \cdots \\
\vdots & \vdots & \vdots \\
\cdots &
X_{i, 1} f_{1, t_N}^{\alpha} + X_{i, 2} f_{2, t_N}^{\alpha} + \dots + X_{i, 6} f_{6, t_N}^{\alpha} +
\sum X_{i, Industry} \text{Industry Factors}_{t_N} +
\sum X_{i, Style} \text{Style Factors}_{t_N} & \cdots \\
\end{bmatrix}
$$

and use the signal that does not vary across equity but time to specify $g(f_1, f_2, \dots, f_6, t)$.

```python
fac_model_ret = {}

for date in frames:
    fac_model_ret[date] = get_estu(frames[date])[facret[date].index.values] * facret[date].T
```

We eventually have to express this signal in terms of the six alpha factors, $f_1, \dots, f_6$.

Holding the belief that this independent component that varies across time but not across equities
is a proxy to the composite alpha factor that we want to find, we may perform regression tree to predict
the ICA 1 values with the six alpha factors' returns. Using regression tree would enable us to explore
non-linear combinations of the six factors' returns (in hopes to closely predict the independent component)

We will use the data from 2003 to 2005 as training data and the 2006 - 2007 data as validation data to tune
the hyper-parameters.

```python
def objective(trial:Trial,train=None,y=None,val=None,y_val=None):

    params = {
        "objective": "regression",
        "verbosity": -1,
        "boosting_type": "gbdt",
        "n_estimators":trial.suggest_int('n_estimators',1,6,step=1),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-8, 10.0, log=True),
        "num_leaves": trial.suggest_int("num_leaves", 5, 25),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf',400,2400,step=400),
        'learning_rate': trial.suggest_float('learning_rate',0.01,0.2,step=0.01)
    }

    gb_regg = LGBMRegressor(**params)
    gb_regg.fit(train,y)
    preds = gb_regg.predict(val)
    mse = np.mean((y_val - preds)**2)

    return mse
```

# Problem 2.

Referring to formula (4.3) for the Markowitz portfolio, and the covariance matrix in (4.29), code up a function to
compute the Markowitz portfolio for each date in our sample. Refer to the helpful code above to get the different pieces
in (4.29). For the risk-aversion constant, use $\kappa$ = 1e-5. Restrict yourself to the estimation universe that was
used above in get_estu. Use your composite alpha factor from Problem 1 as the substitute for $ \mathbb{E}[\mathbf{r}]$.
It is recommended to use the fast inversion formula (4.32) to speed things up. Compute the dot product of your portfolio
with the return, that is compute $\mathbf{h} \cdot \mathbf{r}$ for each date in the sample, and plot the cumulative sum
of the results. For $\mathbf{r}$, use the column called "Ret" in the same data frame that was used to compute the
portfolio itself.

The Markowitz portfolio maximize the mean-variance objective that is
$$
\underset{h\in \mathbb{R}^n}{max}(\textbf{h}' \mathbb{E}[\textbf{r} - \frac{1}{2}\kappa \textbf{h}‘\Sigma\textbf{h}])\\
\Sigma = \mathbb{V}[\textbf{r}] \in S^n_{++}
$$
By solving the first-order condition, we can gain the portfolio weights:
$$
\textbf{h}^* = (\kappa\Sigma)^{-1}\mathbb{E}[\textbf{r}]
$$
where
$$
\Sigma = XFX’ + D\\
$$
If we apply the Woodbury matrix-inversion lemma on $\Sigma$,
$$
\Sigma^{-1} = D^{-1} - D^{-1}X(F^{-1}+X’D^{-1}X)^{-1}X’D^{-1}\\
(\kappa \Sigma)^{-1} = \frac{1}{\kappa}\Sigma^{-1}  = \frac{1}{\kappa} \left[D^{-1} -
D^{-1}X(F^{-1}+X’D^{-1}X)^{-1}X’D^{-1}\right]
$$
Thus, we can get the daily porfolio which is the daily number of holdings for each stock with the daily expected returns
$\mathbb{E}[\textbf{r}]$.

That is,

$$
\begin{aligned}
\textbf{h}^* &= \frac{1}{\kappa}\Sigma^{-1} \\
&= \frac{1}{\kappa} \left[D^{-1} - D^{-1}X(F^{-1}+X’D^{-1}X)^{-1}X’D^{-1}\right] \mathbb{E}[\textbf{r}]\\
&= \frac{1}{\kappa} \left[D^{-1} - D^{-1}X(F^{-1}+X’D^{-1}X)^{-1}X’D^{-1}\right]\cdot \beta^{composite} \cdot g(f_1,f_2,\cdots,f_6)
\end{aligned}
$$

```python
port_returns_l = []
port_weights_l = []
long_returns_l = []
short_returns_l = []

for dt in my_dates_test:

    estu = get_estu(frames_test[dt.strftime("%Y%m%d")])
    n = estu.shape[0]
    estu['Ret'] = wins(estu['Ret'], -0.25, 0.25)

    expect_return = ica_returns_pred.loc[dt,estu.index]

    rske = risk_exposures(estu)
    F = diagonal_factor_cov(dt.strftime("%Y%m%d"), rske)
    X = np.asarray(rske)
    D = np.asarray( (estu['SpecRisk'] / (100 * math.sqrt(252))) ** 2 )
    kappa = 1e-5

    D_inv =  np.diag(1/D)
    F_inv = np.diag(1/np.diag(F))
    Cov_inv = D_inv - D_inv.dot(X).dot(np.linalg.inv(F_inv + X.T.dot(D_inv).dot(X))).dot(X.T).dot(D_inv)

    # Calculate the portfolio weights
    h = (1/kappa)*(Cov_inv.dot(expect_return))
    port_weights_l.append(h)

    # long position weights
    long = h.copy()
    long[long<0] = 0
    # short position weights
    short = h.copy()
    short[short>0] = 0

    long_ret = long.dot(estu['Ret'])
    short_ret = short.dot(estu['Ret'])
    long_returns_l.append(long_ret)
    short_returns_l.append(short_ret)

    # Calcualte the daily portfolio returns
    portfolio_ret = h.dot(estu['Ret'])
    port_returns_l.append(portfolio_ret)
```

The Markowitz optimized porfolio weights for the dates of 2008 to 2010 are shown below.
(It is important to note that the estimation universe does not always filter the same stocks every day.
Hence, some of the equities' weights on certain days are shown as `NAN`.)
